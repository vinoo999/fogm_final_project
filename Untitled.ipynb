```python
zeisel_cells = pd.read_csv('./data/Zeisel.txt', sep="\t", quotechar='"', nrows=10, header=None)
zeisel_cells = zeisel_cells.set_index(0)
zeisel_counts = pd.read_csv('./data/Zeisel.txt', sep="\t", quotechar='"', skiprows=lambda x: x in range(1,10), header=0)
zeisel_counts = zeisel_counts.drop(axis=1, columns='cell_id')
zeisel_counts.head()
```

```python
print(np.unique(zeisel_cells.loc['level1class']))
```

```python
microglia = np.where(zeisel_cells.loc['level1class'] == 'microglia')[0]
oligodendrocytes = np.where(zeisel_cells.loc['level1class'] == 'oligodendrocytes')[0]
interneurons = np.where(zeisel_cells.loc['level1class'] == 'interneurons')[0]
print(oligodendrocytes.shape)
print(interneurons.shape)
print(microglia.shape)
idxs = np.concatenate((microglia, interneurons))
reduced_mat = zeisel_counts.iloc[:,idxs]

log_reduced = np.log2(1+np.array(reduced_mat.values.T, dtype=np.float32))
```

```python
ae = np.where(zeisel_cells.loc['level1class'] == 'astrocytes_ependymal')[0]
em = np.where(zeisel_cells.loc['level1class'] == 'endothelial-mural')[0]
pc = np.where(zeisel_cells.loc['level1class'] == 'pyramidal CA1')[0]
ps = np.where(zeisel_cells.loc['level1class'] == 'pyramidal SS')[0]
print(ae.shape)
print(em.shape)
print(pc.shape)
print(ps.shape)
```
